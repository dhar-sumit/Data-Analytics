## Chi-Squared Test

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('healthcare_dataset.csv')
df.head()
```

```python
df = df.drop(columns=['id'])
df = df.dropna()
df.isna().sum()
```

```python
data_categorical = df.select_dtypes(exclude=['number']).columns.tolist()

encoder = LabelEncoder()
for col in data_categorical:
    df[col] = encoder.fit_transform(df[col])

df.head()
```

```python
x = df.drop(columns=['heart_disease'])
y = df['heart_disease']

chi2_selector = SelectKBest(chi2, k=5)
chi2_selector.fit(x, y)

features = x.columns[chi2_selector.get_support()]
print("The top features are:", list(features))
```

## Anova Test

```python
import pandas as pd
from sklearn.feature_selection import f_classif
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('healthcare_dataset.csv')
df.head()
```

```python
df = df.drop(columns=['id'])
df['bmi'] = df['bmi'].fillna(df['bmi'].median())
df.head()
```

```python
df_categorical = df.select_dtypes(exclude=['number']).columns.tolist()

encoder = LabelEncoder()
for col in df_categorical:
    df[col] = encoder.fit_transform(df[col])

df.head()
```

```python
x = df.drop(columns=['stroke'])
y = df['stroke']
f_values, p_values = f_classif(x, y)
anova_res = pd.DataFrame({'Features': x.columns, 'F values': f_values, 'P values': p_values})
anova_res
```

```python
top_feature = anova_res.head().sort_values(by=['F values', 'P values'], ascending=[False, True])
print("Anova Tested top 5 features are:\n", top_feature.to_string(index=False))
```

## PCA

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('healthcare_dataset.csv')
df.head()
```

```python
df = df.drop(columns=['id'])
df['bmi'] = df['bmi'].fillna(df['bmi'].median())
df.head()
```

```python
encoder = LabelEncoder()
for col in df.select_dtypes(exclude=['number']).columns:
    df[col] = encoder.fit_transform(df[col])
df.head()
```

```python
x, y = df.drop(columns=['stroke']), df['stroke']
```

```python
scale = StandardScaler()
x_scaled = scale.fit_transform(x)
x_scaled
```

```python
pca = PCA(n_components=x.shape[1])
pca.fit_transform(x_scaled)
exp_var = pca.explained_variance_ratio_

plt.figure(figsize=(8,5))
plt.plot(np.cumsum(exp_var), marker='o', color='r')
plt.xlabel('No of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Principal Components')
plt.grid()
plt.show()
```

```python
cum_var = np.cumsum(exp_var)
comp = np.argmax(cum_var >= 0.95)+1

pca_final = PCA(n_components=comp)
x_pca_final = pca_final.fit_transform(x_scaled)
x_pca_final
```

```python
df_pca = pd.DataFrame(x_pca_final, columns=[f"PC{i+1}" for i in range(comp)])
df_pca['stroke'] = y
df_pca.head()
```

## Gini index and decision tree.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
x = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(criterion='gini', random_state=42)
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1:.2f}')
```

```python
plt.figure(figsize=(10,8))
plot_tree(clf, filled=True, feature_names=x.columns, class_names=data.target_names, rounded=True)
plt.show()
```

```python
# Importing required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Loading the Iris dataset
data = load_iris()

# Create a DataFrame from the dataset
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creating the Decision Tree classifier with 'entropy' criterion (Information Gain)
clf = DecisionTreeClassifier(criterion='entropy', random_state=42)

# Fitting the model to the training data
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Calculating the metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Displaying the performance metrics
print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1:.2f}')

# Plotting the Decision Tree
plt.figure(figsize=(12,8))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=data.target_names, rounded=True)
plt.show()
```
